````python
page = requests.get(url).content
hq_soup = bs(page, 'lxml')
earthquake = hq_soup.find('tbody', {'id':'tbody'}).find_all('tr')

new_earthquakes = []

for elem in earthquake[:21]:
    
    date, time = elem.find('td', {'class':'tabev6'}).find('a').text.split()
    lat, lon = elem.findAll('td', {'class':'tabev1'})
    d1, d2, mag = elem.findAll('td', {'class':'tabev2'})
    
    lat_deg = lat.text.strip()
    lon_dg = lon.text.strip()
    dir1 = d1.text.strip()
    dir2 = d2.text.strip()
    mag = mag.text.strip()
    
    region = elem.find('td', {'class':'tb_region'}).text.strip().lower().title()
    
    new_earthquakes.append([date, time, lat_deg, dir1, lon_dg, dir2, mag, region])
    

df = pd.DataFrame(new_earthquakes, columns = ['Date','Time','Lat','Lat Dir','Long','Long Dir', 'Magnitude','Region'])
df
#here I needed help because I wans't able to figure it out on my own
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
req = requests.get(url).content
soup = BeautifulSoup(req, 'lxml')

native = []
native_lang = []

for line in soup.findAll('tr'):
    for l in line.findAll('td'):
        native.append(l.text.strip())

for text in native:
    text = text.strip()
    native_lang.append(" ".join(re.findall("[a-zA-Z]+", text)))

native_lang = list(filter(None, native_lang))
native_lang
df = pd.DataFrame(native_lang[:21:2], columns=['top_10_languages'])
df
````
